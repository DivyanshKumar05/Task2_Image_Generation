```python
import torch
from diffusers import StableDiffusionPipeline

print("Downloading and loading the model... (This takes a minute)")

# 1. Define the 'pipe' variable by loading the model
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)

print("Model loaded! Moving to GPU...")

# 2. Now that 'pipe' exists, move it to the GPU
pipe = pipe.to("cuda")

print("Success! You are ready to generate images.")
```

```python
prompts = [
    A highly detailed futuristic city at sunset, neon lights, cyberpunk style, digital art, masterpiece",
    "A massive futuristic sports stadium at night, packed crowd, glowing holographic scoreboards, dramatic floodlights, hyper-realistic, wide angle shot, 8k resolution, cinematic atmosphere, masterpiece",
    "An astronaut riding a horse on Mars",
    "A well defined picture of a Samurai With his sword in his hand"'-j
    ]
negative_prompt = "blurry, low quality, distorted, deformed,ugly, duplicate, morbid, mutilated, extra fingers, mutated hands, poorly drawn hands, poorly drawn face, mutation, deformed, cloned face, disfigured, gross proportions"
```

```python
import matplotlib.pyplot as plt
```

```python
for i, prompt in enumerate(prompts):

    print(f"\nGenerating Image {i+1}")
    print("Prompt:", prompt)

    image = pipe(prompt).images[0]

    # Save image
    filename = f"generated_image_{i+1}.png"
    image.save(filename)

    # Show image
    plt.figure(figsize=(8,8))
    plt.imshow(image)
    plt.axis("off")
    plt.show()

print("Task completed successfully.")
```
